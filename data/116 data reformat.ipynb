```python
import pandas as pd
```

```python
df2020 = pd.read_csv('GSE_2020.csv')
```

```python
df2020['date'] = pd.to_datetime(df2020['service_date'])
```

```python
df2020.head
```

```python
# Add a new column with the month name
df2020['month'] = df2020['date'].dt.strftime('%B')  # Month name (e.g., 'January')
df2020.head
```

```python
# Group by 'station_name' and 'month_year', and apply aggregation (e.g., sum of scores)
grouped2020 = df2020.groupby(['station_name', 'month']).agg({
    'gated_entries': 'sum',  # Sum the 'score' values for each group
    'route_or_line' : 'first'
    
}).reset_index()
```

```python
grouped2020.head
grouped2020['year'] = 2020
```

```python
grouped2020.to_csv("2020_data.csv", index=False)
```

```python
df2021 = pd.read_csv('GSE_2021.csv')

df2021['date'] = pd.to_datetime(df2021['service_date'])



# Add a new column with the month name
df2021['month'] = df2021['date'].dt.strftime('%B')  # Month name (e.g., 'January')


# Group by 'station_name' and 'month_year', and apply aggregation (e.g., sum of scores)
grouped2021 = df2021.groupby(['station_name', 'month']).agg({
    'gated_entries': 'sum',  # Sum the 'score' values for each group
    'route_or_line' : 'first'
    
}).reset_index()

grouped2021.head
grouped2021['year'] = 2021

grouped2021.to_csv("2021_data.csv", index=False)
```

```python
csv_files = ['2023_data.csv','2022_data.csv','2021_data.csv','2020_data.csv','2019_data.csv','2018_data.csv',
            '2017_data.csv',
            '2016_data.csv',
            '2015_data.csv',
            '2014_data.csv',]
# Create an empty list to store DataFrames
dfs = []

# Loop through the list of files and read each one into a DataFrame
for file in csv_files:
    df = pd.read_csv(file)  # Read each CSV into a DataFrame
    dfs.append(df)  # Append the DataFrame to the list

# Concatenate all DataFrames into one
df_combined = pd.concat(dfs, ignore_index=True)

# Save the combined DataFrame to a new CSV file
df_combined.to_csv('2014-2022_entryData.csv', index=False)

# Print the first few rows of the combined DataFrame
print(df_combined.head())
```
